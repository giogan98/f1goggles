Comparazione tramite sovrapposizione della traccia delle velocità di due giri in pista.

```python
import matplotlib.pyplot as plt
import fastf1.plotting


fastf1.Cache.enable_cache('cache')

# enable some matplotlib patches for plotting timedelta values and load
# FastF1's default color scheme
fastf1.plotting.setup_mpl()

# load a session and its telemetry data
session = fastf1.get_session(2022, 'Bahrain', 'Q')
session.load()
session.event['EventName']
```

Per prima cosa dobbiamo selezionare i due giri di nostro interesse:

```python
lec_lap = session.laps.pick_driver('LEC').pick_fastest()
ver_lap = session.laps.pick_driver('VER').pick_fastest()
```

Una volta fatto questo, dobbiamo ottenere i dati telemetrici per ciascun giro. Aggiungiamo inoltre la 'Distance', per rendere più facile il confronto fra i due giri.

```python
lec_tel = lec_lap.get_car_data().add_distance()
ver_tel = ver_lap.get_car_data().add_distance()
```

Infine, possiamo creare il grafico e rappresentare entrambe le linee di velocità. I colori di queste saranno scelti in base al team di appartenenza dei piloti.

```python
fer_color = fastf1.plotting.team_color('FER')
rbr_color = fastf1.plotting.team_color('RBR')

fig, ax = plt.subplots()
ax.plot(lec_tel['Distance'], lec_tel['Speed'], color=fer_color, label='LEC')
ax.plot(ver_tel['Distance'], ver_tel['Speed'], color=rbr_color, label='VER')

ax.set_xlabel('Distance in m')
ax.set_ylabel('Speed in km/h')

ax.legend()
plt.suptitle(f"Fastest Lap Comparison \n "
             f"{session.event['EventName']} {session.event.year} Qualifying")

plt.show()
```
